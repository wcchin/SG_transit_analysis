# 2.4 bus stops

```python

url = "http://datamall2.mytransport.sg/ltaodataservice/BusStops"
response = requests.get(
    url,
    params={'q': 'requests+language:python'},
    headers={'AccountKey': 'dYrbZaCHRaWZrgEZIE4q2A==', 
             'Accept': 'application/json'},
)
json_response = response.json()
```

```python
len(json_response["value"])
```

```python
json_response["value"]
```

```python
for i in range(10):
    print(i*500+500)
```

```python
res = []
for i in range(11):
    url = "http://datamall2.mytransport.sg/ltaodataservice/BusStops?"
    if i>0:
        url = url + "$skip={}".format(str(i*500))
    response = requests.get(
        url,
        params={'q': 'requests+language:python'},
        headers={'AccountKey': 'dYrbZaCHRaWZrgEZIE4q2A==', 
                 'Accept': 'application/json'},
    )
    json_res = response.json()
    res.append(json_res)
```

```python
count = 0
for jres in res:
    count+=(len(jres["value"]))
    #break
count
```

```python
busstops = []
for jres in res:
    busstops.extend(jres["value"])
len(busstops)
```

```python
df_busstop = pd.DataFrame.from_dict(busstops)
len(df_busstop)
```
